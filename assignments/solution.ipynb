```python
# Implementing the Fibonacci Sequence

def fibonacci(a: int, b: int, limit: int):
    """Print Fibonacci numbers up to a given limit, one per line."""
    while a <= limit:
        print(a)
        a, b = b, a + b

fibonacci(0, 1, 5)
```

```python
# Finding the Greatest Common Divisor (GCD) using the Euclidean Algorithm

def find_gcd(a: int, b: int):
    """Return the greatest common divisor of a and b using the Euclidean algorithm."""
    while b:
        a, b = b, a % b
    return a
print(find_gcd(48, 60))
```

```python
# Finding the Greatest Common Divisor (GCD) using Subtraction method

def find_gcd(a: int, b: int):
    """Return the greatest common divisor of a and b using the Subtraction method."""
    while a != b:
        if a >= b:
            a -= b
        else:
            b -= a
    return a
print(find_gcd(54, 24))
```

```python
# Implementing Fibonacci Sequence using Recursion

def find_fib_rec(n):
    """Return the nth Fibonacci number using recursion."""
    if n < 2:
        return n
    else:
        return find_fib_rec(n-1) + find_fib_rec(n-2)

def find_fib(n):
    """Print first n Fibonacci numbers"""
    for i in range(n):
        print(find_fib_rec(i))
find_fib(5)
```

```python

# Functions, Lists, Loops, and Conditionals
import random
def filter_by_threshoild(numbers, threshold):
    """Return a list of numbers greater than the given threshold."""
    new_list = []
    for number in numbers:
        if number >= threshold:
            new_list.append(number)
        else:
            print(f"{number} is below the threshold")
    return new_list

random_numbers = [random.randint(-100, 100) for _ in range(10)]
print("Random Numbers:", random_numbers)
filtered_numbers = filter_by_threshoild(random_numbers, 0)
print("Filtered Numbers:", filtered_numbers)
```

```python
# Dictionaries and list enumeration

data = [
    [19, "apple", 23, "banana"],
    ["orange", 42, "mango", 66],
    [11, "cat", "kiwi", 82]
]
string_dict = {}
integer_dict = {}
for row_index, row in enumerate(data):
    for col_index, item in enumerate(row):
        if type(item) == str:
            string_dict[(row_index, col_index)] = item
        elif type(item) == int:
            integer_dict[(row_index, col_index)] = item
string_dict, integer_dict
```

```python
# Finding Prime Numbers up to a Given Limit

primes = []
limit = 100
for num in range(2, limit + 1):
    is_prime = True
    for divisor in range(2, num):
        if num % divisor == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)
print(primes)
```

```python
# Matplotlib for Visualization

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("../data/employment-12211-9008_en.csv", header=0, index_col=None)
year = 2015
data_year = data[data['Year'] == year]

sectors = data_year['Sector']
male = data_year['Male']
female = data_year['Female']

x = np.arange(len(sectors))
width = 0.4

plt.figure(figsize=(12, 8))
plt.bar(x - width/2, male, width, label='Male', color='steelblue')
plt.bar(x + width/2, female, width, label='Female', color='lightcoral')

plt.xlabel("Sector")
plt.ylabel("Employment")
plt.title(f"Male vs Female Employment by Sector in {year}")
plt.xticks(x, sectors, rotation=90, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Stacked chart comparison

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../data/employment-12211-9008_en.csv", header=0, index_col=2)

pivot_data = data.pivot_table(values="Total", index="Year", columns="Sector", aggfunc="sum")
pivot_data.plot(kind="bar", stacked=True, figsize=(12, 6))

plt.title("Total Employment by Sector per Year")
plt.xlabel("Year")
plt.ylabel("Total Employment")
plt.legend(title="Sector", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
